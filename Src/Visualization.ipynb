```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_csv("../Data/df1.csv")
df2 = pd.read_csv("../Data/df2.csv")
df3 = pd.read_csv("../Data/df3.csv")
```

# Noble awarded by country (top 10)


```python
var_1=df1["Country"].value_counts().head(10)

plot_1=sns.countplot(y= df1["Country"], order=var_1.index, palette = "tab10")
plot_1.bar_label(plot_1.containers[0])
plot_1.set(title = 'Number of Nobel Prizes')
plot_1.set(xlabel='Number of prizes')
plt.figure(figsize = (15,6))

plot_1.figure.savefig("../images/Top10_nobels.jpg", dpi=1000, bbox_inches = "tight")
```

# Top 7 countries by Nobel Prizes by gender

```python
df1 = pd.read_csv("../Data/df1.csv")
var_2=df1["Country"].value_counts().head(7)

plot_2=sns.countplot(data=df1, y = df1["Country"], hue = "gender", palette = "tab10", order=var_2.index)
plot_2.bar_label(plot_2.containers[0])
plot_2.bar_label(plot_2.containers[1])
plot_2.set(title = 'Number of Nobel Prizes by gender')
plot_2.set(xlabel='Number of prizes')
plt.figure(figsize = (15,6))

plot_2.figure.savefig("../images/Top10_nobels_gender.jpg", dpi=1000, bbox_inches = "tight")
```

# Category of the Nobel Prizes for the top 5 awarded

```python
# plot_4_size, axs = plt.subplots(nrows = 1, ncols = 1, figsize = (40,30))
var_2=df1["Country"].value_counts().head(5)

plot_4=sns.countplot(data=df1, y = df1["Country"], hue = "Category", palette = "tab10", order=var_2.index)
plot_4.bar_label(plot_4.containers[0])
plot_4.bar_label(plot_4.containers[1])
plot_4.bar_label(plot_4.containers[2])
plot_4.bar_label(plot_4.containers[3])
plot_4.bar_label(plot_4.containers[4])
plot_4.bar_label(plot_4.containers[5])
plot_4.set_title('Category of Nobel Prizes', fontsize=20)
plot_4.set(xlabel='Number of prizes')
plt.figure(figsize = (15,12))
```
