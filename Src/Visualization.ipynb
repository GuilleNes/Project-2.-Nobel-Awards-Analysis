# Noble awarded by country (top 10)


```python
var_1=df1["Country"].value_counts().head(10)

plot_1=sns.countplot(y= df1["Country"], order=var_1.index, palette = "tab10")
plot_1.bar_label(plot_1.containers[0])
plot_1.set(title = 'Number of Nobel Prizes')
plot_1.set(xlabel='Number of prizes')
plt.figure(figsize = (15,6))

plot_1.figure.savefig("../images/Top10_nobels.jpg", dpi=1000, bbox_inches = "tight")
```

# Top 7 countries by Nobel Prizes by gender

```python
df1 = pd.read_csv("../Data/df1.csv")
var_2=df1["Country"].value_counts().head(7)

plot_2=sns.countplot(data=df1, y = df1["Country"], hue = "gender", palette = "tab10", order=var_2.index)
plot_2.bar_label(plot_2.containers[0])
plot_2.bar_label(plot_2.containers[1])
plot_2.set(title = 'Number of Nobel Prizes by gender')
plot_2.set(xlabel='Number of prizes')
plt.figure(figsize = (15,14))

plot_2.figure.savefig("../images/Top10_nobels_gender.jpg", dpi=1000, bbox_inches = "tight")
```

# Evolution on the number of awarded along the time

```python
df_1 = df1[(df1["Country"]=="USA")]
df_2 = df1[(df1["Country"]=="United Kingdom")]
df_3 = df1[(df1["Country"]=="Germany")]
df_4 = df1[(df1["Country"]=="France")]
df_5 = df1[(df1["Country"]=="Sweden")]
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(25,20))

sns.histplot(data=df_1, x=df1.AwardYear, hue="gender", ax=axs[0], bins=40).set_title('USA',x=0.1, y = 0.8, fontsize = 16)
sns.histplot(data=df_2, x=df1.AwardYear, hue="gender", ax=axs[1], bins=40).set_title('United Kingdom', x=0.1, y = 0.8, fontsize = 16)
sns.histplot(data=df_3, x=df1.AwardYear, hue="gender", ax=axs[2], bins=40).set_title('Germany', x=0.1, y = 0.8, fontsize = 16)
sns.histplot(data=df_4, x=df1.AwardYear, hue="gender", ax=axs[3], bins=40).set_title('France', x=0.3, y = 0.8, fontsize = 16)
sns.histplot(data=df_5, x=df1.AwardYear, hue="gender", ax=axs[4], bins=40).set_title('Sweden', x=0.1, y = 0.8, fontsize = 16)
```
