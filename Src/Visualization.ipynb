```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

# Noble awarded by country (top 10)


```python
var_1=df1["Country"].value_counts().head(10)

plot_1=sns.countplot(y= df1["Country"], order=var_1.index, palette = "tab10")
plot_1.bar_label(plot_1.containers[0])
plot_1.set(title = 'Number of Nobel Prizes')
plot_1.set(xlabel='Number of prizes')
sns.set(rc={'figure.figsize':(14.0,8.5)})
plot_1.figure.savefig("../images/Plot_1.jpg", dpi=1000, bbox_inches = "tight")
```

# Top 7 countries by Nobel Prizes by gender

```python
df1 = pd.read_csv("../Data/df1.csv")
var_2=df1["Country"].value_counts().head(7)

plot_2=sns.countplot(data=df1, y = df1["Country"], hue = "gender", palette = "tab10", order=var_2.index)
plot_2.bar_label(plot_2.containers[0])
plot_2.bar_label(plot_2.containers[1])
plot_2.set(title = 'Number of Nobel Prizes by gender')
plot_2.set(xlabel='Number of prizes')
sns.set(rc={'figure.figsize':(14.0,8.5)})

plot_2.figure.savefig("../images/Plot_2.jpg", dpi=1000, bbox_inches = "tight")
```

```python
df1_ = df1[(df1["AwardYear"]<2020)]
plot_3 = sns.kdeplot(x=df1.AwardYear, hue=df1.gender, fill=True)
sns.set(rc={'figure.figsize':(14.0,8.5)})
plot_3.figure.savefig("../images/plot_3.jpg", dpi=1000, bbox_inches = "tight")


```

# Evolution on the number of awarded along the time

```python
df_1 = df1[(df1["Country"]=="USA")]
df_2 = df1[(df1["Country"]=="United Kingdom")]
df_3 = df1[(df1["Country"]=="Germany")]
df_4 = df1[(df1["Country"]=="France")]
df_5 = df1[(df1["Country"]=="Sweden")]
fig, axs = plt.subplots(nrows=5, ncols=1, figsize=(25,20))

plot_5 = sns.histplot(data=df_1, x=df1.AwardYear, hue="gender", multiple="stack", ax=axs[0], bins=40).set_title('USA',x=0.1, y = 0.8, fontsize = 16)
plot_6 = sns.histplot(data=df_2, x=df1.AwardYear, hue="gender", multiple="stack", ax=axs[1], bins=40).set_title('United Kingdom', x=0.1, y = 0.8, fontsize = 16)
plot_7 = sns.histplot(data=df_3, x=df1.AwardYear, hue="gender", multiple="stack", ax=axs[2], bins=40).set_title('Germany', x=0.1, y = 0.8, fontsize = 16)
plot_8 = sns.histplot(data=df_4, x=df1.AwardYear, hue="gender", multiple="stack", ax=axs[3], bins=40).set_title('France', x=0.3, y = 0.8, fontsize = 16)
plot_9 = sns.histplot(data=df_5, x=df1.AwardYear, hue="gender", multiple="stack", ax=axs[4], bins=40).set_title('Sweden', x=0.1, y = 0.8, fontsize = 16)

plot_5.figure.savefig("../images/plot_5.jpg", dpi=1000, bbox_inches = "tight")
```

# Category of the Nobel Prizes for the top 5 awarded

```python
# plot_4_size, axs = plt.subplots(nrows = 1, ncols = 1, figsize = (40,30))
var_2=df1["Country"].value_counts().head(5)

plot_10=sns.countplot(data=df1, y = df1["Country"], hue = "Category", palette = "tab10", order=var_2.index)
plot_10.bar_label(plot_10.containers[0])
plot_10.bar_label(plot_10.containers[1])
plot_10.bar_label(plot_10.containers[2])
plot_10.bar_label(plot_10.containers[3])
plot_10.bar_label(plot_10.containers[4])
plot_10.bar_label(plot_10.containers[5])
plot_10.set_title('Category of Nobel Prizes', fontsize=20)
plot_10.set(xlabel='Number of prizes')
sns.set(rc={'figure.figsize':(14.0,8.5)})
plot_10.figure.savefig("../images/plot_10.jpg", dpi=800, bbox_inches = "tight")
```

On the plot above, we can see how the percentage of literature or peace awarded in countries like USA or UK is much smaller than others like France or Sweden
